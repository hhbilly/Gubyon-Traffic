```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

hourly = pd.read_csv('data/hourly.csv', date_parser='time')
hourly
```

```python
hourly.plot(x='time', y=range(1,9), kind='line', title='Speed by time of day')
plt.show()
```

```python
hourly.plot(x='time', y=range(4,9), kind='line', title='Counts of cars exceeding speed limit by time of day')
plt.savefig('cesl.png')
```

```python
hourly.sum()[4:9].sum()
```

```python
hourly.sum()[1:9].plot.bar(title='Counts in speed ranges')
plt.savefig('cisr.png')
```

```python
pct2float = lambda x:float(x.strip('%'))
daily = pd.read_csv('data/daily.csv', date_parser='time', converters={2:pct2float,4:pct2float,6:pct2float})
daily
```

```python
daily.plot(x='Day', y=[1,3,5], kind='bar', title='Counts over speed limit')
plt.savefig('cosl.png')
```

```python
daily.plot(x='Day', y=[2,4,6], kind='bar', title='Percentages over speed limit')
plt.savefig('posl.png')
```
